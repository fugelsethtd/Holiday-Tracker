```python
#importing
import json
import datetime
from bs4 import BeautifulSoup
import requests

#decorator
def deletesuccess(func):
    def printdelmessage(delholiday):
        print('Holiday has been deleted')
    return printdelmessage
    

#creating the holiday class
class Holiday:
    '''Holiday class includes a name and a date for holidays in the us'''
    def __init__(self, name, date):
        self.__name = name
        self.__date = date
    def __str__(self):
        return ('{} ({})'.format(self.__name,self.__date))
    @deletesuccess
    def __del__(self):
        print('Destructor Called')
    @property
    def name(self):
        return self.__name
    @name.setter
    def name(self, new_name):
        self.__name = new_name
    @property
    def date(self):
        return self.__date
    @date.setter
    def date(self, new_date):
        self.__date = new_date
```

```python
#import holidays from the given json
holidayobjectlist = []
with open(r'C:/Users/slygu/Downloads/holidays.json') as jsonfile:
    data = json.load(jsonfile)
    for i in range(0,len(data['holidays'])):
        #print(data['holidays'][i]['name'])
        #print(data['holidays'][i]['date'])
        year = int(data['holidays'][i]['date'][0:4])
        month = int(data['holidays'][i]['date'][5:7])
        day = int(data['holidays'][i]['date'][8:10])
        newholiday = Holiday(data['holidays'][i]['name'], datetime.date(year, month, day))
        holidayobjectlist.append(newholiday)
for i in range(0, len(holidayobjectlist)):
    print(holidayobjectlist[i])
```

```python
#get holidays for 2020
html = requests.get('https://www.timeanddate.com/holidays/us/2020')
soup = BeautifulSoup(html.text,'html.parser')
#print(soup.find('table').prettify())
table = soup.find('table')


holidaylist = []
for row in table.find_all_next('tr'):
    cells = row.find_all_next('td')
    datehead = row.find_all_next('th')
    if len(cells) > 1 and len(datehead) > 1:
        holidayscrape = {}
        holidayscrape['name'] = cells[1].string
        holidayscrape['date'] = datehead[0].string
        holidaylist.append(holidayscrape)
#print(holiday2021list)

for i in range(0, len(holidaylist)):
    if holidaylist[i]['date'] != 'Date':
        gooddate = datetime.datetime.strptime(holidaylist[i]['date'][0:3] , '%b')
        spaceindex = holidaylist[i]['date'].index(' ')
        day = holidaylist[i]['date'][spaceindex:]
        #print(gooddate.month)
        dateobject = datetime.date(2020, gooddate.month, int(day))
        newholiday = Holiday(holidaylist[i]['name'], dateobject)
        holidayobjectlist.append(newholiday)
for i in range(0, len(holidayobjectlist)):
    print(holidayobjectlist[i])
```

```python
## creating the menu
menu = True
nosave = True
while menu:
    print('Holiday Menu')
    print('=============')
    print('1. Add a Holiday')
    print('2. Remove a Holiday')
    print('3. Save Holiday List')
    print('4. View Holidays')
    print('5. Exit')
    menuoption = int(input('Please enter your menu option(1-5): '))
    if menuoption == 1:
        print('Add a Holiday')
        print('=============')
        holidayname = str(input('Holiday: '))
        date = str(input('Date (YYYY-MM-DD): '))
        try:
            year = int(date[0:4])
            month = int(date[5:7])
            day = int(date[8:10])
            newholiday = Holiday(holidayname, datetime.date(year, month, day))
            holidayobjectlist.append(newholiday)
            print('The following holiday has been added:')
            print(newholiday)
        except:
            print('Date not formatted correctly, please try again')
            
    if menuoption == 2:
        print('Remove a Holiday')
        print('================')
        holidayremove = str(input('Holiday to be deleted: '))
        removal = False
        for i in range(0, len(holidayobjectlist)):
            if holidayobjectlist[i].name == holidayremove:
                holidayobjectlist.pop(i)
                removal = True
        if removal:
            print('{} has been removed.'.format(holidayremove))
        else:
            print('{} was not found on list'.format(holidayremove))
    if menuoption == 3:
        print('Save Holiday List')
        print('=================')
        saveoption = str(input('Are you sure you want to save your changes? [y/n]: '))
        if saveoption.lower() == 'y':
            with open(r'C:\Users\slygu\Documents\Python week 2\holidayslist.json', 'w') as outfile:
                savelist = []
                for i in range(0,len(holidayobjectlist)):
                    savedict = {}
                    savedict['name'] = holidayobjectlist[i].name
                    savedict['date'] = str(holidayobjectlist[i].date)
                    #print(savedict)
                    savelist.append(savedict)
                json.dump(savelist, outfile, indent=4)
                nosave = False
    if menuoption == 4:
        print('View Holidays')
        print('=============')
        yearoption = int(input('Which year? [2020-2024]: '))
        weekoption = str(input('Which week? [1-52, leave blank for the current week]: '))
        if weekoption != '':    
            yearfiltered = list(filter(lambda x: int(x.date.year) ==  yearoption, holidayobjectlist))
            weekfiltered = list(filter(lambda x: int(x.date.isocalendar()[1]) == int(weekoption), yearfiltered))
            for i in range(0, len(weekfiltered)):
                print(weekfiltered[i])
            
        else:
            todayweek = datetime.date.today().isocalendar()[1]
            yearfiltered = list(filter(lambda x: int(x.date.year) ==  2021, holidayobjectlist))
            weekfiltered = list(filter(lambda x: int(x.date.isocalendar()[1]) == todayweek, yearfiltered))
            for i in range(0, len(weekfiltered)):
                print(weekfiltered[i])
            weatheroption = str(input('Would you like to see the weather for the week? [y/n]:'))
            if weatheroption == 'y':
                print(weatherdict)
        
    if menuoption == 5:
        print('Exit')
        print('====')
        if nosave:
            print('Are you sure you want to exit?')
            print('Unsaved changes will be lost')
            exitoption = str(input('[y/n]: '))
            if exitoption.lower() == 'y':
                print('Goodbye!')
                break
        else:
            exitoption = str(input('Are you sure you want to exit? [y/n]: '))
            if exitoption.lower() == 'y':
                print('Goodbye!')
                break
    else:
        print('Please enter number 1-5')
print('Thank you for using the holiday menu!')    
```
